```python
from networkx.algorithms import tree
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
weights = [random.randint(1, 100) for i in range(30)]
```

```python
grafo = nx.Graph()

grafo.add_edge("A","B", weight = weights[0])
grafo.add_edge("A","C", weight = weights[1])
grafo.add_edge("A","D", weight = weights[2])
grafo.add_edge("B","C", weight = weights[3])
grafo.add_edge("B","D", weight = weights[4])
grafo.add_edge("B","E", weight = weights[5])
grafo.add_edge("C","D", weight = weights[6])
grafo.add_edge("C","E", weight = weights[7])
grafo.add_edge("C","F", weight = weights[8])
grafo.add_edge("D","E", weight = weights[9])
grafo.add_edge("D","F", weight = weights[10])
grafo.add_edge("D","G", weight = weights[11])
grafo.add_edge("E","F", weight = weights[12])
grafo.add_edge("E","G", weight = weights[13])
grafo.add_edge("E","H", weight = weights[14])
grafo.add_edge("F","G", weight = weights[15])
grafo.add_edge("F","H", weight = weights[16])
grafo.add_edge("F","I", weight = weights[17])
grafo.add_edge("G","H", weight = weights[18])
grafo.add_edge("G","I", weight = weights[19])
grafo.add_edge("G","J", weight = weights[20])
grafo.add_edge("H","I", weight = weights[21])
grafo.add_edge("H","J", weight = weights[22])
grafo.add_edge("H","A", weight = weights[23])
grafo.add_edge("I","J", weight = weights[24])
grafo.add_edge("I","A", weight = weights[25])
grafo.add_edge("I","B", weight = weights[26])
grafo.add_edge("J","A", weight = weights[27])
grafo.add_edge("J","B", weight = weights[28])
grafo.add_edge("J","C", weight = weights[29])

pos=nx.spring_layout(grafo)
labels=nx.get_edge_attributes(grafo,'weight')

plt.figure()
nx.draw_networkx_edge_labels(grafo, pos, edge_labels = labels)
print("Nodos del grafo original: ", grafo.nodes())
print("Aristas del grafo original: ", grafo.edges())
nx.draw(grafo, pos, with_labels = True,node_color = "skyblue")
plt.savefig("GrafoOriginal.png",dpi=300)
```

```python
spanning_tree = tree.minimum_spanning_tree(grafo, algorithm = "kruskal")
labels=nx.get_edge_attributes(spanning_tree,'weight')
nx.draw_networkx_edge_labels(spanning_tree, pos, edge_labels =labels)
print("Nodos del arbol de expansión: ", spanning_tree.nodes())
print("Aristas del arbol de expansión: ", spanning_tree.edges())
nx.draw(spanning_tree,pos,with_labels=True, node_color = "violet")
plt.savefig("ArbolExpansión.png",dpi=300)
```
